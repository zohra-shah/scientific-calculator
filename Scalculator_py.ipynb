```python
# First, install necessary libraries
!pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt

# Scientific Calculator Functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b

def power(a, b):
    return a ** b

def sqrt(a):
    return np.sqrt(a)

# Function for plotting
def plot_function(expr, x_range):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = eval(expr)
    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    plt.title(f'Plot of {expr}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    plt.show()

# Example usage
print("Scientific Calculator Operations:")
print(f"Add: {add(5, 3)}")
print(f"Subtract: {subtract(5, 3)}")
print(f"Multiply: {multiply(5, 3)}")
print(f"Divide: {divide(5, 0)}")  # Division by zero
print(f"Power: {power(5, 3)}")
print(f"Sqrt: {sqrt(16)}")

# Plotting example
print("Plotting y = x^2")
plot_function('x**2', (-10, 10))
```
